```python
# Combine into a function
files = glob("MetaMotion/*.csv")

def read_data_from_file(files):
    # Extracting features from rest of the filenames
    acc_df = pd.DataFrame()
    gyro_df = pd.DataFrame()

    acc_set = 1
    gyro_set = 1

    for f in files:
        # Add metadata columns extracted from the filename:
        participant = os.path.basename(f).split("-")[0]

        label = os.path.basename(f).split("-")[1]

        category = os.path.basename(f).split("-")[2].rstrip("123").rstrip("_MetaWear_2019")

        # Displaying the f dataframe
        df = pd.read_csv(f)

        # Add metadata columns extracted from the filename:
        df['participant'] = participant
        df['label'] = label
        df['category'] = category

        if 'Accelerometer' in f:
            df["set"] = acc_set
            acc_set += 1
            acc_df = pd.concat([acc_df,df])
        
        if 'Gyroscope' in f:
            df["set"] = gyro_set
            gyro_set += 1
            gyro_df = pd.concat([gyro_df,df])

    acc_df.index = pd.to_datetime(acc_df['epoch (ms)'] ,unit='ms') # replacing the index with converted epoch value
    gyro_df.index = pd.to_datetime(gyro_df['epoch (ms)'] ,unit='ms') # replacing the index with converted epoch value
            
    # Removing the time columns from the accelometer dataframe
    del acc_df["epoch (ms)"]
    del acc_df["time (01:00)"]
    del acc_df["elapsed (s)"]

    # Removing the time columns from the gyroscope dataframe
    del gyro_df["epoch (ms)"]
    del gyro_df["time (01:00)"]
    del gyro_df["elapsed (s)"]

    return acc_df, gyro_df

acc_df, gyro_df = read_data_from_file(files)
```

```python
# Merging the accelometer and gyrocscope dataframe together
merged_data = pd.concat([acc_df.iloc[:,:3],gyro_df],axis=1)
merged_data.head(20)
```

```python
# Renaming the columns in the merged_data dataframe
merged_data.columns = [
    "acc_x",
    "acc_y",
    "acc_z",
    "gyro_x",
    "gyro_y",
    "gyro_z",
    "participant",
    "label",
    "category",
    "set"
]

# Displaying the first five instances of the dataframe
merged_data.head()
```

```python
# Resampling the dataframe
# The gryroscope (25hz) run twice as fast as the accelerometer(12.5hz)
sampling = {
    "acc_x":"mean",
    "acc_y":"mean",
    "acc_z":"mean",
    "gyro_x":"mean",
    "gyro_y":"mean",
    "gyro_z":"mean",
    "participant":"last",
    "label":"last",
    "category":"last",
    "set":"last"
}

resampled = merged_data.resample("200ms").apply(sampling)

# Split by day
days = (g for n,g in merged_data.groupby(pd.Grouper(freq='D')))
data_resampled = pd.concat([df.resample(rule="200ms").apply(sampling).dropna() for df in days])

data_resampled
```
